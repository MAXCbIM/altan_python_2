# Matplotlib


```python
import numpy as np
import matplotlib.pyplot as plt
```

- простые линейные графики

```python
# 1-мерный массив (вектор) из 10 значений 
x = np.arange(10)
x
```

```python
y = x
```

```python
# визиуализация
plt.plot(x, y)
plt.show()
```

```python
y = x ** 2

plt.plot(x, y)
plt.show()
```

```python
# 1-мерный массив из 1000 значений
# линейная последовательность в диапазоне от 0 до 10
x = np.linspace(0, 10, 1000)
x
```

```python
# данные по оси х
x = np.linspace(-10, 10, 1000)

# определение функции
def sigmoid(arg):
    return 1 / (1 + np.exp(-arg))

# вычисление значений по оси y
y = sigmoid(x)

# визиуализация
plt.plot(x, y)
plt.show()
```

```python
# гиперболический тангенс
y = np.tanh(x)

plt.plot(x, y)
plt.show()
```

```python
# x = np.linspace(-10, 10, 10)
x = np.linspace(-10, 10, 1000)

y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# совмещение нескольких графиков в одном окне
x = np.linspace(-10, 10, 1000)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = sigmoid(x)
y4 = np.tanh(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.show()
```

```python
# цвет линий

plt.plot(x, y1, color="red") # по названию
plt.plot(x, y2, color="m") # по первой букве цвета из форматов "rgb" "cmyk"
plt.plot(x, y3, color="#ffa70350", lw=6) # по 16-ричному коду формата (RGBa), lw-пиксель (тольшина) 
plt.plot(x, y4, color=(0.0, 1.0, 0.5)) # по формату RGB в виде кортежа со значениями в диапазоне от 0 до 1
plt.show()
```

```python
# стиль линий

plt.plot(x, np.sin(x), linestyle='-', color="k", lw=5) #"solid") # сплошная 
plt.plot(x, np.sin(x - 1), linestyle='--', color="b", lw=3) #"dashed") # штриховая 
plt.plot(x, np.sin(x - 2), linestyle=':', color="g", lw=4) #"?") # пунктирная 
plt.plot(x, np.sin(x - 3), linestyle='-.', color="r", lw=2) #"?") # штрихпунктирная 
plt.show()
```

```python
# обьединение настроек цвета и стиля линий

plt.plot(x, np.sin(x), "--g", lw=2)
plt.plot(x, np.sin(x - 1), ":b", lw=2)
plt.plot(x, np.sin(x - 2), "-.k", lw=2)
plt.plot(x, np.sin(x - 3), "-y", lw=2)
plt.show()
```

```python
# пределы осей

plt.plot(x, np.sin(x), "--g", lw=2) 

# индивидуальные пределы
# plt.xlim(-15, 15)
# plt.ylim(-1.5, 1.5)

plt.axis([-20, 20, -1.5, 1.5])
plt.show()
```

```python
# легенда графика 

plt.plot(x, np.sin(x), "--b", lw=1, label="Синус") 
plt.plot(x, np.cos(x), ":r", lw=2, label="Косинус") 

plt.legend()

plt.axis([-20, 20, -2, 2])

plt.show()
```

```python
# подписи осей
plt.plot(x, np.sin(x), "--b", lw=1, label="Синус") 
plt.plot(x, np.cos(x), ":r", lw=2, label="Косинус") 

plt.legend()
plt.xlabel("x")
plt.ylabel("sin and cos")

plt.axis([-20, 20, -2, 2])

plt.show()
```

```python
# ЗФГОЛОВОК ГРАФИКА
plt.plot(x, np.sin(x), "--b", lw=1, label="Синус") 
plt.plot(x, np.cos(x), ":r", lw=2, label="Косинус") 

plt.legend()
plt.xlabel("x")
plt.ylabel("sin and cos")
plt.title("Графики синуса и косинуса")

plt.axis([-20, 20, -2, 2])

plt.show()
```
