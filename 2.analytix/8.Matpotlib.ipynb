```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# интерактивный режим
# %matplotlib notebook
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x))
plt.show()
```

```python
# настройка габаритов окна
plt.figure(figsize=(8, 5)) #, facecolor="red", frameon=False)
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.figure(figsize=(8, 5))
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x), label="синусоида")
plt.plot(x, np.cos(x), label="косинусоида")
# настройка легенды
plt.legend(loc="upper right", labelcolor="green", frameon=True)
# сетка
plt.grid(True)
plt.show()
```
